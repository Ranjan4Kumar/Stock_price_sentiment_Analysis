```python
# Problem statement Stock sentiment analysis using News Headline
"""
=> The dataset in consideration is a combination of the world news and stock proce shift available on kaggale
=> There are 25 columns of the top news headlines for each day in the data frame
=> Data ranges from 2008 to 2016 and the data from 2000 to 2008 was scrapped from  yahoo finance
=> Labels are absed on the Dow Jones Industrial Average stock index
Class 1 -> the stock price increased
class 0 -> The stock price stayed the same or dcreased
"""
from google.colab import drive
drive.mount('/content/drive/')

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/DS/Stock_Price_sentiment_analysis/Data.csv', encoding = "ISO-8859-1")

# encoding -> 
```

```python
df.shape
```

```python
df.head(1)
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
# Taking all the feature column from the table that is from 2nd to 27th column because 2nnd column is target column whihc is lable

df1 = train.iloc[:,2:27]

# Removing all the punctuations from the table here it means that apart from small a-z and A-Z replace everything from the table including punctuations and all
df1.replace("[^a-zA-Z]" , " " , regex = True , inplace = True)
```

```python
# Renaming columns names according to my convienence

list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
df1.columns = new_index
df1.head(1)
```

```python
# Converting all the Capital words to small words

for index in new_index:
  df1[index] = df1[index].str.lower()
df1.head(1)
```

```python
# For applying NLP we have to add all the headlines in a single para then only we can add the values to it for that what we are doing is that.

headlines = [] # we created a empty list over here
for row in range(0, len(df1.index)):
  headlines.append(' '.join(str(x) for x in df1.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# Now implememt BOW

cv = CountVectorizer(ngram_range = (2,2)) #
traindataset = cv.fit_transform(headlines)
```

```python
# Implement RandomForest classifier

randomclassifier = RandomForestClassifier(n_estimators=200 , criterion = 'entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for thr test dataset
df_test = test.iloc[: ,2:27]

# removing all the punctuations
df_test.replace("[^a-zA-z]", " ", regex = True , inplace = True)
```

```python
# changing all the column name to numerical form
list2= [i for i in range(25)]
new_index2 = [str(i) for i in list2]
df_test.columns = new_index2
df_test.head(1)
```

```python
for index in new_index2:
  df_test[index] = df_test[index].str.lower()
df_test.head(1)
```

```python
# For applying NLP we have to add all the headlines in a single para then only we can add the values to it for that what we are doing is that.
headline2 = []
for row1 in range(0, len(df_test.index)):
  headline2.append(' '.join(str(x) for x in df_test.iloc[row1 ,0:25]))
```

```python
headline2[0]
```

```python
test_dataset = cv.transform(headline2)
prediction = randomclassifier.predict(test_dataset)
```

```python
# check for accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

matrix = confusion_matrix(test['Label'] , prediction)
print(matrix)

```

```python
score = accuracy_score(test['Label'],prediction)
print("%.2f"%(score*100),'%')
```

```python
report = classification_report(test['Label'] ,prediction)
print(report)
```
